```python
let_28_found_in_L = []

for i in range(len(qTxCons)-27):
    let_28 = []
    j = i
    while len(let_28) < 28 and j < len(qTxCons):
        rasm = arbCons2RasmD[qTxCons[j]]
        if rasm not in let_28:
            let_28.append(rasm)
        j += 1
    let_28_found_in_L.append(j-i+1)
```

```python
avg_28_let_found_in = sum(let_28_found_in_L) / len(let_28_found_in_L)
```

```python
import numpy as np
from scipy import stats as st
min(let_28_found_in_L), st.mode(let_28_found_in_L), np.median(let_28_found_in_L), avg_28_let_found_in, max(let_28_found_in_L)
```
